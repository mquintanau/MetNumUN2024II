```
from sympy import Symbol, div

# Define la variable simbólica
x = Symbol('x')

# Define el polinomio
p = (
    x**10 - 67*x**9 + 1970*x**8 - 33430*x**7 + 362033*x**6
    - 2609971*x**5 + 12660220*x**4 - 40708020*x**3
    + 82816416*x**2 - 95954112*x + 47900160
)

# Realiza divisiones secuenciales desde (x-1) hasta (x-9)
for i in range(1, 10):  # Itera de 1 a 9
    print(f"Dividiendo por (x-{i}):")
    cociente, residuo = div(p, x-i)  # Realiza la división
    print(f"Cociente: {cociente}")
    print(f"Residuo: {residuo}\n")

    # Actualiza el polinomio con el cociente para la próxima iteración
    p = cociente
```

## SymPy’s Symbolic and Numercical Complex Evaluations
- https://docs.sympy.org/latest/modules/evalf.html](https://)
- x = x1 + I*x2,y = y1 + I*y2, z = z1 + I*z2, x * y * z

<img src="https://raw.githubusercontent.com/gjhernandezp/NM/main/SymPyLab/sympylabwolfram7.jpg" />

```
from sympy import symbols, I, N, pi, expand
x1, x2, y1, y2, z1, z2 = symbols("x1 x2 y1 y2 z1 z2", real=True)
x = x1 + I*x2
y = y1 + I*y2
z = z1 + I*z2

print(x*y*z)
print(expand(x*y*z))
print(expand((x*y)*z))
print(expand(x*(y*z)))
w=1/(pi + I)
print('w=',w)
w=N(w)
print('w=',w)
```
